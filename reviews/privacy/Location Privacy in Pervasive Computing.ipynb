## The solution proposed 

Privacy-protection framework based on frequently changing pseudonyms so users avoid being identified by the locations they visit by:
 - the concept of mix zones maps location privacy to anonymous communications
 - hide the true identity from the application receiving the location
 - compartmentalization by letting location services register for location callbacks only at the same site where the service is provided
 
### Hypothesis
1. A middleware event model through which entities entering or exiting a predefined region of space generates events. Applications register their interrest in a particular set of locations and locatables and receive callbacks when the corresponding events occurred 
2. The middleware is trusted and may help users to hide their identity
3. Users registers interest in particular application with the middleware
4. Application receives event callbacks from the middleware when user enter, moves or exits the application-defined area
5. Each update period, the middleware determines wheter any event have occured and issues callbacks  to application when appropriate
6. Availability of an  anonimyzation proxy for all communications between users and application 



 
### Definitions

- David Chaum proposes for anonymous communication:
   - Mix network
   - Dining cryptographers algorithm
   - Anonymity sets
   
- Mix network: a store-and-forward network that offers anonymous communication facilities:
     - an observer can not trace a message from its source to its destination without the collusion of the mix nodes

- Mix node: 
    - collects $n$ equal-length packets as input and reorders them by some metric before forwarding them, thus providing unlinkability between incoming and outgoing messages
    - can encrypt the messages
    - the anonymity set is the number of distinct senders in the batch

## Mix zones 

- A mix zone is :  
  - for a group of user a connected spatial region of maximum size in which none of these users has registered any application callback. For a given group of users there migth be several distinct mix zone
  - analogous to mix node
  

- An application zone is the area where a user has a registered a callback for.

- The middleware can define the mix zone apriori or calculate them separatly for each group of users as the spatial areas currently not in an application zone 

- Application do not receive any information when user are in the mix zone (identities are mixed)

- Assuming users change to a new and unsuded pseudonym whenever they enter a mix zone, applications that see a user emerging from the mix zone cannot distinguish that  a user from any other who was in the mix zone at the same time and cannot link people going into the mix zone with those coming out of it

- If a mix zone has a diameter $(D)$ much larger than the distance $(d(u,m))$ the user can cover during one location update period, it migth not mix users adequatly. For example the figutre below :

### Observation on Anonymity sets

- Anonymity sets is only a good measure of anonymity when all the members of the set are equally likely to be the one of interest to the observer (maxiaml entropy)

- For a set of $n$ elements equiprobable, we need 
$\log_2(n)$

bits of information to identify one of them. 

- When not equiprobale, fewer bits can be use as follow:
   - Consider the movement of a user in a mix zone $z$ at $t$
   - For user crossing $z$, the preceding zone $p$ can be recorded and the time $t-1$ 
   - $s$ is the zone visited at $t+1$
   - If the user does not change a zone $z=p=s$
   - Historical data can be used to determine the relative frequencies for each user in $z$ the pair $(p,s)$
 
 - A movement Matrix $M$:
   - records for all possible $(p,s)$ pairs the number of times a person who was in $z$ at $t$ was in $p$ at $t-1$ and in $s$ in $t+1$
   
 - Therefore the joint probability can be determined :
   
   $P(prec=p, subs=s) = \frac{M(p,s)}{\sum_{i,j}{M(i,j}}$
 
 - The conditional probaility :
 
   $P(prec=p | subs=s) = \frac{M(p,s)}{\sum_{j}{M(p,j}}$
   
 - The entropy :
 
 $h = - \sum p_i \dot \log p_i$
 
 

```python
# Generation of data frame containing the values given in the paper
import pandas as pd
import numpy as np

# Location Matrix collected for one Mix  Zone and 4 application zones
# The location update period is 5s
# The users do not often renturn to the same mix zonet they came from
# It is unlikely that that $p=s$ in every case except a prolonged stay in the mix zone.
values = [
    [96, 47, 66, 101, 814],
    [125, 13, 17, 1, 43],
    [29,55,7, 22, 30],
    [39,6,64,39,66],
    [24,47,74,176,162]
]

#E: East, N:North, S:South, W:West, Z:Mix Zone
labels = ['E', 'N', 'S', 'W', 'Z']
index =  labels[::-1]

# Create an DataFrame 
df = pd.DataFrame(values,index =index, columns=labels)


def joint_proba(df, x, y):
    """Determines the joint probability of x and y"""
    return df[x][y]/df.values.sum()

def cond_proba(df, x,y):
    """Determines the conditional probability of x given y"""
    return df[x][y]/df[y].values.sum()

def entropy(probas):
    """Determines the entropy for given probabilities"""
    e = 0
    for p in probas:
        e += p*np.log2(p)
        
    return -e

print(df)
```

### Impact of user movement model on anonymity in the mix zone

- Given two user:
    - walking in opposite directions (one east to west, one west to east)
    - they pass throug the empty mix zone in the same time period
    - An observer knows from the application zones that the users went to mix zone and the directions they came from
    - The observer can know that the user came out of the mix zone with new pseudonyms
    - The observer want to link  the pseudonyms (find out whether they both went straigth ahead or each made a uturn)
         - Without any apriori knowledge, the information content of this alternative is one full bit
         - By using the movement matrix, the observer can guess the answer with a chance of success better than random
    

#### Numerical example
 - Given :
    - the two users moving from p and s to z
    - if $p$ and $s$ are from ${E,W}$, the Matrix can be reduced to $2x2$ Matrix
    - There are 16 cases for the users: EEEE, EEEW, ..., WWWW
    - EEEE means (User 1 came from E to E and the second user form E to E)
    

```python
# Extract the 2x2 matrix

def sub_dataframe(df, rows, cols):
    return df.loc[rows,cols]

m = sub_dataframe(df,['W','E'],['E', 'W'])

m
```

##### Given :
- the obeserved movement of the two users 

    $observed = EEWW \vee EWEW \vee WEEW \vee WWEE$
    
- the uturn which corresponds to both the users doind a uturn
    
     $uturn = EEWW \vee WWEE$
     
- Where $\vee$ is the logical or

- if $P(XXXX) = P(XX) \times P(XX)$ then:
    
    - $P(observed) = \sum P(XXXX)$


```python
# List of the observed movement by an adversary
observed = ['EEWW', 'EWWE', 'WEEW', 'WWEE']

# List of the uturn
uturn = ['EEWW', 'WWEE']


def proba(movements):
    p = 0
    
    for mvt in movements:
        if (len(mvt) %2 == 0) :
        
            p = p + joint_proba(m,mvt[0:2][0],mvt[0:2][1])*joint_proba(m,mvt[2:4][0],mvt[2:4][1])
    return p

p_observed = proba(observed)
p_uturn = proba(uturn)

print('The probability of observed is : %.3f ' % p_observed )
print('The probability of uturn is : %.4f ' % p_uturn)
# The conditiona probability of P(u|o) = P(u /\ o) /p(o)
# as u is in o, u /\ o = u 
print('The conditional probability of uturn given observed : %.3f ' % (p_uturn/p_observed))

print('The entropy is : %.3f ' % entropy ([0.001]))
```

#### Conclusions
- Given the probability of $uturn = 0.001$, the observer knows for $0.999$ that the users went straigth forward
 and thus can link the pseudonyms.
